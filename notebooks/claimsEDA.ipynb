```python
import pandas as pd

beneficiary2008 = pd.read_csv(
    "../data/raw/beneficiary/176613_DE1_0_2008_Beneficiary_Summary_File_Sample_14/DE1_0_2008_Beneficiary_Summary_File_Sample_14.csv"
)

beneficiary2009 = pd.read_csv(
    "../data/raw/beneficiary/176584_DE1_0_2009_Beneficiary_Summary_File_Sample_14/DE1_0_2009_Beneficiary_Summary_File_Sample_14.csv"
)

beneficiary2010 = pd.read_csv(
    "../data/raw/beneficiary/176598_DE1_0_2010_Beneficiary_Summary_File_Sample_14/DE1_0_2010_Beneficiary_Summary_File_Sample_14.csv"
)

beneficiary = pd.concat([beneficiary2008, beneficiary2009, beneficiary2010], ignore_index=True)

inpatient = pd.read_csv(
    "../data/raw/inpatient//176533_DE1_0_2008_to_2010_Inpatient_Claims_Sample_14/DE1_0_2008_to_2010_Inpatient_Claims_Sample_14.csv"
)
```

```python
datasets = [beneficiary, inpatient]

for dataset in datasets:
    dt_columns = [col for col in dataset.columns if col.endswith('_DT')]
    for col in dt_columns:
        dataset[col] = pd.to_datetime(dataset[col], errors='coerce')
```
